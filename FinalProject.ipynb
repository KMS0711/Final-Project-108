```python
def clean_time(time):
    return time.split()[0]
```

```python
df["duration"] = df["duration"].apply(clean_time)
```

```python
#remove columns name, country, category, backers, goal, and pledged
# we remove backers because that is only known after the fact, and we want to be able to predict success of a campaign
# we will use usd_goal_real and usd_pledged_real because the conversions have already been done to USD for these
```

```python
df = df.drop(['name', 'country', 'category', 'backers', 'goal', 'pledged', 'usd pledged','currency'], axis=1)
```

```python
df.head()
```

```python
#rename columns for simplicity
#note that we are using names that previously meant similar things but were not as useful; don't be confused.
df.rename(columns={'usd_pledged_real':'pledged',
                          'usd_goal_real':'goal',
                          'main_category': 'category'
                          'Unnamed'}, inplace = True)
df
```

```python
df.state.unique()
```

```python
#remove rows with states other than successful, failed, or cancel, and change canceled to failed
df = df.drop(df[df.state == 'live'].index)
df = df.drop(df[df.state == 'undefined'].index)
df = df.drop(df[df.state == 'suspended'].index)
```

```python
df
```

```python
def cancel_to_fail(state):
    if state == 'canceled':
        state = 'failed'
    return state
```

```python
df['state'] = df['state'].apply(cancel_to_fail)
df
```
